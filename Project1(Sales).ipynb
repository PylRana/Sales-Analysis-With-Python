# Sales Analysis 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Diwali Sales Data.csv",encoding="unicode-escape")
```

```python
# Deleting Status & Unnamed column
df=df.drop(columns=['Status','unnamed1'])
# df.drop(['Status','unnamed1'],axis=1,inplace=True) Another method using axis
```

```python
df
```

```python
# DELETING NANS
df.dropna(inplace=True)
```

```python
# To change the data type of amount from float to int
df['Amount']=df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.columns
```

```python
# TO CHANGE NAME OF COLUMN OR COLUMNS
df.rename(columns={'Marital_Status':'Status'},inplace=True)
```

## Exploratory Data Analysis

```python
plt.figure(figsize=(5,5))
x=sns.countplot(data=df,x='Gender')
plt.title("Gender Wise Sales") # Adding title to the chart
#x.bar_label(x.containers[0]) # for labeling
# method 2 for labelling
for bars in x.containers:
    x.bar_label(bars)
plt.show()
```

 ### From the Above Graph , We can clearly conclude that there are more female customers than male customers.

### Marital Status Wise Sale

```python
plt.figure(figsize=(5,5))
ms=sns.countplot(data=df,x='Status')
ms.bar_label(ms.containers[0])
plt.title("Marital Status Wise Sales")
plt.xlabel("Marital Status")
plt.ylabel("Sales")
plt.show()
```

### From above Bar Chart, It is evident that married customers tend to shop more than unmarried customers.

```python
status_sale=df.groupby(['Status','Gender'], as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False)
plt.figure(figsize=(5,5))
sns.barplot(data=status_sale,x='Status',y='Amount',hue='Gender')
```

### From the Above Bar chart, We can conclude that Most of the married buyers are female with high purchasing power.

```python
# Group By
plt.figure(figsize=(5,5))
sales_gender=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gender)
```

### from above graphs, We can analyse that females have more purchasing power than male i.e., Female customers spend more on shopping.

```python
c=sns.countplot(data=df,x='Age Group',hue='Gender')
# for bars in c.containers:
#     c.bar_label(bars)
c.bar_label(c.containers[0])
```

```python
sales_age=df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=sales_age,x='Age Group',y='Amount')
```

 ### From the above bar graph, we can analyze that majority of buyers fall within the age group of 26-35 , especially females

```python
state_sales=df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=state_sales,x='State',y='Orders')
```

```python
state_sales=df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=state_sales,x='State',y='Amount')
```

### From Above we can say that most of the orders and total sales_amount are from UP,Maharastra,Karnataka,respectively

```python
# plt.figure(figsize=(18,5))
sns.set(rc={'figure.figsize':(18,5)})
s=sns.countplot(data=df,x='Occupation')
# s.bar_label(s.container[0])
for bars in s.containers:
    s.bar_label(bars)
```

```python
occp_sales=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(data=occp_sales,x='Occupation',y='Amount')
```

### From above Chart, It is evident that most of the buyers are working in IT, Healthcare and Aviation.

```python
# plt.figure(figsize=(18,5))
sns.set(rc={'figure.figsize':(20,5)})
z=sns.countplot(data=df,x='Product_Category')
# s.bar_label(s.container[0])
for bars in z.containers:
    z.bar_label(bars)
```

```python
cat_sales=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.barplot(data=cat_sales,x='Product_Category',y='Amount')
```

### From above insights, we can analyse that most of the sold product are food,clothing & Electronics

# Conclusion

### In Conclusion, the sales analysis reveals that married women, particulary aged between 25-35 show a strong inclination towards shopping compared to the male counterparts.
### It signifies that we have a strong market foothold in the U.p.,Maharastra, Karnataka as well as the IT sector,Healthcare & Aviation industries.
